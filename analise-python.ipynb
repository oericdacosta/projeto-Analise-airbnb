14. Desenhe um mapa da cidade de Nova York, localizando somente os top 10
im√≥veis com mais avalia√ß√µes, por regi√£o.

```python
brooklyn = df.loc[df['neighbourhood_group'] == 'Brooklyn', :].sort_values(by='number_of_reviews', ascending=False).head(10)
manhattan = df.loc[df['neighbourhood_group'] == 'Manhattan', :].sort_values(by='number_of_reviews', ascending=False).head(10)
queens = df.loc[df['neighbourhood_group'] == 'Queens', :].sort_values(by='number_of_reviews', ascending=False).head(10)
island = df.loc[df['neighbourhood_group'] == 'Staten Island', :].sort_values(by='number_of_reviews', ascending=False).head(10)
bronx = df.loc[df['neighbourhood_group'] == 'Bronx', :].sort_values(by='number_of_reviews', ascending=False).head(10)

df_aux = pd.concat([brooklyn, manhattan, queens, island, bronx], ignore_index=True)

df_aux = df_aux.loc[:, ['latitude', 'longitude', 'number_of_reviews']]

map = folium.Map()

marker_cluster = MarkerCluster().add_to(map)

for index, info in df_aux.iterrows():

    folium.Marker(location=[info['latitude'], info['longitude']],
                icon=folium.Icon(icon='glyphicon glyphicon-cutlery'),
                ).add_to(marker_cluster)

map
```

16. Desenhe um mapa da cidade de Nova York, localizando somente os top 10
menores pre√ßos, por regi√£o.


```python
df_aux = df.loc[df['price'] != 0, :]

brooklyn = df_aux.loc[df_aux['neighbourhood_group'] == 'Brooklyn', :].sort_values(by='price', ascending=True).head(10)
manhattan = df_aux.loc[df_aux['neighbourhood_group'] == 'Manhattan', :].sort_values(by='price', ascending=True).head(10)
queens = df_aux.loc[df_aux['neighbourhood_group'] == 'Queens', :].sort_values(by='price', ascending=True).head(10)
island = df_aux.loc[df_aux['neighbourhood_group'] == 'Staten Island', :].sort_values(by='price', ascending=True).head(10)
bronx = df_aux.loc[df_aux['neighbourhood_group'] == 'Bronx', :].sort_values(by='price', ascending=True).head(10)

df_aux = pd.concat([brooklyn, manhattan, queens, island, bronx], ignore_index=True)

df_aux = df_aux.loc[:, ['latitude', 'longitude', 'price']]

map = folium.Map()

marker_cluster = MarkerCluster().add_to(map)

for index, info in df_aux.iterrows():

    folium.Marker(location=[info['latitude'], info['longitude']],
                icon=folium.Icon(icon='glyphicon glyphicon-cutlery'),
                ).add_to(marker_cluster)

map
```

17. Desenhe um gr√°fico de barras, mostrando a m√©dia de aluguel di√°ria de cada regi√£o.

```python
df_aux = df[['price', 'neighbourhood_group']].groupby(['neighbourhood_group']).mean().reset_index()
fig = px.bar(df_aux, x='neighbourhood_group', y='price', title = 'Valor m√©dio do aluguel di√°rio por regi√£o (NYC)')

fig.update_xaxes(title_text='Regi√µes')
fig.update_yaxes(title_text='Pre√ßo')

fig.show()
```

18. Desenhe um gr√°fico de barras, mostrando a mediana de aluguel di√°ria de
cada regi√£o.


```python
df_aux = df[['price', 'neighbourhood_group']].groupby(['neighbourhood_group']).median().reset_index()
fig = px.bar(df_aux, x='neighbourhood_group', y='price', title = 'Mediana do aluguel di√°rio por regi√£o (NYC)')

fig.update_xaxes(title_text='Regi√µes')
fig.update_yaxes(title_text='Pre√ßo')

fig.show()
```

19. Desenhe um gr√°fico de barras, mostrando a soma de todos os valores da
di√°ria de cada regi√£o.


```python
df_aux = df[['price', 'neighbourhood_group']].groupby(['neighbourhood_group']).sum().reset_index()
fig = px.bar(df_aux, x='neighbourhood_group', y='price', title = 'Soma dos valores por regi√£o (NYC)')

fig.update_xaxes(title_text='Regi√µes')
fig.update_yaxes(title_text='Valor Total')

fig.show()
```

20. Desenhe um gr√°fico de barras, mostrando o menor valor de di√°ria de cada
